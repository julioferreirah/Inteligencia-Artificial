### Preparação dos dados

```python
# Informações DataFrame
data.info()
# Head
data.head()
```

De cara, vemos que algumas tuplas não tem os 891 dados completos, como:


*   Idade: info importante + pouca/média missing data ▶ Preencher com média
*   Embarque: dado relevante + pouca missing data ▶ excluir tuplas sem
*   Cabine: dado pouco importante + muita missing data ▶ excluir coluna

Também podemos perceber alguns dados inúteis para uma análise como:
*   Nome ▶ Já existe um identifcador
*   Ticket ▶ O código parece não identificar nenhum padrão específico

Temos atributos a serem ajustados:
*  Sexo ▶ Precisa ser codificado
*  Embarcado ▶ Precisa ser codificado
*  Classe ▶ Talvez até poderia ser usado assim devido a ordenação, mas achei melhor codificar tb

E atributos que precisam ser colocas em escala para não super influenciar o modelo:
*  Idade
*  Tarifa

```python
# preencher idade
m = data['Age'].mean()
data['Age'].fillna(m, inplace=True)
```

```python
#tirar Cabine
data.drop('Cabin', axis=1, inplace=True)
```

```python
#tirar Nome
data.drop('Name', axis=1, inplace=True)
```

```python
#tirar Ticket
data.drop('Ticket', axis=1, inplace=True)
```

```python
#tirar tuplas sem idade
data.dropna(axis=0, inplace=True)
```

```python
# codificando Sexo
dummies = pd.get_dummies(data['Sex'])
data = pd.concat([data, dummies], axis = 1)
data.drop('Sex', axis=1, inplace=True)
```

```python
# codificando Embarcado
dummies = pd.get_dummies(data['Embarked'])
data = pd.concat([data, dummies], axis = 1)
data.drop('Embarked', axis=1, inplace=True)
```

```python
# codificando Classe
dummies = pd.get_dummies(data['Pclass'])
data = pd.concat([data, dummies], axis = 1)
data.drop('Pclass', axis=1, inplace=True)
```

```python
# colocando idade em escala
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().fit(data[['Age']])
escala = scaler.transform(data[['Age']])
escala = pd.DataFrame(escala, columns = ['Age'])
data.drop('Age', axis=1, inplace=True)
data = pd.concat([data, escala], axis = 1)
```

```python
# colocando Tarifa em escala
scaler = StandardScaler().fit(data[['Fare']])
escala = scaler.transform(data[['Fare']])
escala = pd.DataFrame(escala, columns = ['Fare'])
data.drop('Fare', axis=1, inplace=True)
data = pd.concat([data, escala], axis = 1)
```

```python
data.drop('PassengerId', axis=1, inplace=True)
```

**Separação de Classe**

> Queremos saber/classificar/predizer se um passageiro morreu ou não com base nos outros dados. Ou seja, se adicionar um novo passageiro, sem o atributo Survived, o algoritmo deve classificado como 0 ou 1 com base nos atributos que aprendeu




```python
data.dropna(axis=0, inplace=True)
data.isna()
```

```python
classe = data['Survived']
data.drop('Survived', axis=1, inplace=True)
```

## Construção de um modelo de classificação


### K-NN

```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=10)
X = data.values
y = classe.values
neigh.fit(X, y)
```

### MPL

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Definição da MLP
model = Sequential()
model.add(Dense(10, input_dim=12, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

# Compilação do modelo
model.compile(loss='mean_squared_error', optimizer='adam')

# Dados de treinamento
X = data.values
y = classe.values

# Treinamento do modelo
model.fit(X, y, epochs=100, batch_size=32)
```

## Avaliação do modelo

Pegando os dados de teste e correção

```python
# importando dados de teste
teste = pd.read_csv("https://raw.githubusercontent.com/hype-usp/PS-2023_1/main/Case/Avancado/data/test.csv")

# importando categorias para verificação
resultado = pd.read_csv("https://raw.githubusercontent.com/hype-usp/PS-2023_1/main/Case/Avancado/data/gender_submission.csv")
```

```python
# tratando dado de teste

m = teste['Age'].mean()
teste['Age'].fillna(m, inplace=True)
teste.drop('Cabin', axis=1, inplace=True)
teste.drop('Name', axis=1, inplace=True)
teste.drop('Ticket', axis=1, inplace=True)
teste.drop('PassengerId', axis=1, inplace=True)
dummies = pd.get_dummies(teste['Sex'])
teste = pd.concat([teste, dummies], axis = 1)
teste.drop('Sex', axis=1, inplace=True)
dummies = pd.get_dummies(teste['Embarked'])
teste = pd.concat([teste, dummies], axis = 1)
teste.drop('Embarked', axis=1, inplace=True)
dummies = pd.get_dummies(teste['Pclass'])
teste = pd.concat([teste, dummies], axis = 1)
teste.drop('Pclass', axis=1, inplace=True)
scaler = StandardScaler().fit(teste[['Age']])
escala = scaler.transform(teste[['Age']])
escala = pd.DataFrame(escala, columns = ['Age'])
teste.drop('Age', axis=1, inplace=True)
teste = pd.concat([teste, escala], axis = 1)
scaler = StandardScaler().fit(teste[['Fare']])
escala = scaler.transform(teste[['Fare']])
escala = pd.DataFrame(escala, columns = ['Fare'])
teste.drop('Fare', axis=1, inplace=True)
teste = pd.concat([teste, escala], axis = 1)
testeN = teste
teste.dropna(axis=0, inplace=True)
teste.info()
teste.head()
```

```python
# Colocando nos modelos
X = teste.values
y_KNN = neigh.predict(X)
print(y_KNN)
```

```python
#tratando resultado esperado
novo = pd.concat([testeN, resultado], axis = 1)
novo.head()
novo.dropna(axis=0, inplace=True)
novo.info()
esperado = novo['Survived'].values
```

```python
# medindo resultado
from sklearn.metrics import accuracy_score
acuracia = accuracy_score(esperado, y_KNN)*100
print("Acurácia: %.2f" %(acuracia) + "%")
```

```python
y_MLP = model.predict(X)
y_MLP = np.round(y_MLP)
print(y_MLP)
acuracia = accuracy_score(esperado, y_MLP)*100
print("Acurácia: %.2f" %(acuracia) + "%")
```
